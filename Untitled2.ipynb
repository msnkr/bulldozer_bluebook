```python
import pandas as pd
import numpy as np

from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error, mean_squared_log_error
from sklearn.model_selection import RandomizedSearchCV
from xgboost import XGBRegressor, XGBRFRegressor

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("data/TrainAndValid.csv", low_memory=False, parse_dates=["saledate"])
```

```python
df.head()
```

```python
df.sort_values(by="saledate", inplace=True)
```

```python
df.isna().sum()
```

```python
encoder = LabelEncoder()

def preprocessing(df):
    df["SaleMonth"] = df.saledate.dt.month
    df["SaleYear"] = df.saledate.dt.year
    df.drop("saledate", axis=1, inplace=True)
    
    for label, content in df.items():
        if pd.api.types.is_numeric_dtype(content):
            if pd.isnull(content).sum():
                df[label] = pd.isnull(content.median())
        else:
            if pd.isnull(content).sum():
                df[label] = df[label].fillna("missing")

            df[label] = encoder.fit_transform(content)
    return df
```

```python
df = preprocessing(df)
```

```python
train_df = df[df.SaleYear != 2012]
valid_df = df[df.SaleYear == 2012]
```

```python
len(train_df), len(valid_df)
```

```python
X_train, y_train = train_df.drop("SalePrice", axis=1), train_df["SalePrice"]
X_valid, y_valid = valid_df.drop("SalePrice", axis=1), valid_df["SalePrice"]
```

```python
bst = XGBRegressor().fit(X_train, y_train)
```

```python
bst.score(X_valid, y_valid)
```

```python
def show_scores(model):
    y_preds_train = model.predict(X_train)
    y_preds_valid = model.predict(X_valid)

    scores = {
        "Test MAE": mean_absolute_error(y_train, y_preds_train),
        "Valid MAE": mean_absolute_error(y_valid, y_preds_valid),
        "TEST MSE": mean_squared_error(y_train, y_preds_train),
        "Valid MSE": mean_squared_error(y_valid, y_preds_valid),
        "Train RMSLE": mean_squared_error(y_train, y_preds_train, squared=False),
        "Valid RMSLE": mean_squared_error(y_valid, y_preds_valid, squared=False),
        # "Train MSLE": mean_squared_log_error(y_train, y_preds_train),
        "Valid MSLE": mean_squared_log_error(y_valid, y_preds_valid),
        "rmsle 2": np.sqrt(mean_squared_log_error(y_valid, y_preds_valid)),
        "Train R2 Score": r2_score(y_train, y_preds_train),
        "Valid R2 Score": r2_score(y_valid, y_preds_valid)
    }

    return scores
```

```python
results = show_scores(bst)
results
```
