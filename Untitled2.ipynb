```python
import pandas as pd
import numpy as np

from sklearn.preprocessing import LabelEncoder
from xgboost import XGBRegressor

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("data/TrainAndValid.csv", low_memory=False, parse_dates=["saledate"])
```

```python
df.head()
```

```python
df.sort_values(by="saledate", inplace=True)
```

```python
df.isna().sum()
```

```python
encoder = LabelEncoder()

def preprocessing(df):
    df["SaleMonth"] = df.saledate.dt.month
    df["SaleYear"] = df.saledate.dt.year
    df.drop("saledate", axis=1, inplace=True)
    
    for label, content in df.items():
        if pd.api.types.is_numeric_dtype(content):
            if pd.isnull(content).sum():
                df[label] = pd.isnull(content.median())
        else:
            if pd.isnull(content).sum():
                df[label] = df[label].fillna("missing")

            df[label] = encoder.fit_transform(content)
    return df
```

```python
df = preprocessing(df)
```

```python
train_df = df[df.SaleYear != 2012]
valid_df = df[df.SaleYear == 2012]
```

```python
len(train_df), len(valid_df)
```

```python
X_train, y_train = train_df.drop("SalePrice", axis=1), train_df["SalePrice"]
X_valid, y_valid = valid_df.drop("SalePrice", axis=1), valid_df["SalePrice"]
```

```python
bst = XGBRegressor().fit(X_train, y_train)
```

```python
bst.score(X_valid, y_valid)
```

```python
_, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=(14, 6))
sns.barplot(data=valid_df, x="SaleMonth", y="SalePrice", ax=ax1, palette="pastel", hue="SaleMonth", legend=False)
sns.barplot(data=train_df, x="SaleMonth", y="SalePrice", ax=ax2, palette="pastel", hue="SaleMonth", legend=False);
```

```python
plt.figure(figsize=(14, 6))
sns.histplot(data=valid_df, x="SalePrice", hue="SaleMonth", kde=True, palette="pastel");
```

```python
plt.figure(figsize=(14, 6))
sns.histplot(data=train_df, x="SalePrice", hue="SaleMonth", kde=True, palette="pastel");
```

```python
test_df = pd.read_csv("data/Test.csv", parse_dates=["saledate"], low_memory=False)
test_df.head()
```

```python
test_df.sort_values(by=["saledate"], inplace=True)
```

```python
len(test_df)
```

```python
test_df = preprocessing(test_df)
test_df.head()
```

```python
y_preds = bst.predict(test_df)
test_df["SalePrice"] = pd.Series(y_preds)
test_df.head()
```

```python
sns.barplot(data=test_df, x="SaleMonth", y="SalePrice", palette="pastel", hue="SaleMonth", legend=False);
```
