```python
import pandas as pd
import numpy as np

from sklearn.preprocessing import LabelEncoder
from xgboost import XGBRegressor

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
train_df = pd.read_csv("data/Train.csv", low_memory=False, parse_dates=["saledate"])
valid_df = pd.read_csv("data/Valid.csv", low_memory=False, parse_dates=["saledate"])
```

```python
train_df.head()
```

```python
train_df.sort_values(by="saledate", inplace=True)
valid_df.sort_values(by="saledate", inplace=True)
```

```python
train_df.isna().sum()
```

```python
encoder = LabelEncoder()

def preprocessing(df):
    df["SaleMonth"] = df.saledate.dt.month
    df["SaleYear"] = df.saledate.dt.year
    df.drop("saledate", axis=1, inplace=True)
    
    for label, content in df.items():
        if pd.api.types.is_numeric_dtype(content):
            if pd.isnull(content).sum():
                df[label] = pd.isnull(content.median())
        else:
            if pd.isnull(content).sum():
                df[label] = df[label].fillna("missing")

            df[label] = encoder.fit_transform(content)
    return df
```

```python
train_df = preprocessing(train_df)
```

```python
train_df.isna().sum()
```

```python
X_train, y_train = train_df.drop("SalePrice", axis=1), train_df["SalePrice"]
```

```python
%%time

bst = XGBRegressor().fit(X_train, y_train)
```

```python
bst.score(X_train, y_train)
```

```python
valid_df = preprocessing(valid_df)
y_valid = bst.predict(valid_df)
```

```python
valid_df["SalePrice"] = pd.Series(y_valid)
```
